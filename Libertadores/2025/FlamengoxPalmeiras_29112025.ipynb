```python
path_flamengo = 'https://raw.githubusercontent.com/LunusMax/football-data/refs/heads/main/Libertadores/2025/df_flamengo.csv'
path_palmeiras = 'https://raw.githubusercontent.com/LunusMax/football-data/refs/heads/main/Libertadores/2025/df_palmeiras.csv'
path_gk = 'https://raw.githubusercontent.com/LunusMax/football-data/refs/heads/main/Libertadores/2025/df_GK.csv'
```

```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import math
from matplotlib import cm
from matplotlib.colors import ListedColormap
from math import pi
from matplotlib.transforms import offset_copy
import matplotlib.patches as patches
from matplotlib.backends.backend_pdf import PdfPages
```

```python
df_flamengo = pd.read_csv(path_flamengo, sep=';')
df_palmeiras = pd.read_csv(path_palmeiras, sep=';')
df_gk = pd.read_csv(path_gk, sep=';')
```

```python
# Criar métricas por 90 para goleiros
df_gk['GA_per90'] = df_gk['GA'] / df_gk['90s']
df_gk['PKatt_per90'] = df_gk['PKatt'] / df_gk['90s']
df_gk['PKsv_per90'] = df_gk['PKsv'] / df_gk['90s']
df_gk['PSxG_per90'] = df_gk['PSxG'] / df_gk['90s']
df_gk['Att_per90'] = df_gk['Att'] / df_gk['90s']
df_gk['Stp_per90'] = df_gk['Stp'] / df_gk['90s']
```

```python
# Criar métricas por 90
cols_per90 = [
    'Gls', 'Ast', 'G+A', 'xG', 'npxG', 'xAG', 'npxG+xAG',
    'Sh', 'SoT',
    'KP', 'SCA', 'GCA', 'PPA', 'CrsPA',
    'PrgC', 'PrgP', 'PrgR',
    'Carries', 'Touches', 'Rec', 'CPA',
    'Tkl', 'TklW', 'Int', 'Blocks', 'Tkl+Int', 'Clr',
    'Won', 'Lost'
]

for col in cols_per90:
    df_flamengo[f'{col}_per90'] = df_flamengo[col] / df_flamengo['90s']
    df_palmeiras[f'{col}_per90'] = df_palmeiras[col] / df_palmeiras['90s']
```

```python
metricas_por_posicao = {
    "GOL": ["Save%", "PK-Save%", "PSxG_per90", "Stp_per90"],
    "LD":  ["PrgC_per90", "PPA_per90", "SCA_per90", "Tkl+Int_per90"],
    "LE":  ["PrgC_per90", "PPA_per90", "SCA_per90", "Tkl+Int_per90"],
    "ZAG": ["Tkl+Int_per90", "Blocks_per90", "Won%", "PrgP_per90"],
    "VOL": ["Tkl+Int_per90", "PrgP_per90", "SCA_per90", "Blocks_per90"],
    "MEI": ["SCA_per90", "GCA_per90", "PrgP_per90", "PPA_per90"],
    "ATA": ["Gls_per90", "xG_per90", "SCA_per90", "PrgC_per90"]
}
```

```python
duelos = {
    "GOL": [
        ("Agustín Rossi", "Carlos Miguel")
    ],

    "LD": [
        ("Guillermo Varela", "Khellven")
    ],

    "ZAG": [
        ("Danilo", "Gustavo Gómez"),
        ("Léo Pereira", "Murilo Cerqueira")

    ],

    "LE": [
        ("Alex Sandro", "Joaquín Piquerez")
    ],

    "VOL": [
        ("Erick Pulgar", "Bruno Fuchs")
    ],

    "MEI": [
        ("Jorginho", "Andreas Pereira"),
        ("Giorgian De Arrascaeta", "Raphael Veiga")
    ],

    "ATA": [
        ("Luiz Araújo", "Allan"),          # Allan atacante do Palmeiras
        ("Jorge Carrascal", "Vitor Roque"),
        ("Bruno Henrique", "José Manuel López")
    ]
}

duelos
```

```python
def gerar_duelo_pdf(df_fla, df_pal, df_gk, jogador_fla, jogador_pal, posicao, metricas, nome_pdf):

    if posicao == "GOL":
        duel = pd.concat([
            df_gk[df_gk["Player"] == jogador_fla],
            df_gk[df_gk["Player"] == jogador_pal]
        ], ignore_index=True)
    else:
        duel = pd.concat([
            df_fla[df_fla["Player"] == jogador_fla],
            df_pal[df_pal["Player"] == jogador_pal]
        ], ignore_index=True)

    fig, axes = plt.subplots(2, 2, figsize=(12.0, 13.5), dpi=100)
    axes = axes.flatten()

    for i, metrica in enumerate(metricas):
        ax = axes[i]

        bars = ax.bar(
            duel["Player"],
            duel[metrica],
            color=["#C00000", "#006437"]
        )

        ax.set_title(
            f"{metrica.replace('_per90','')}",
            fontsize=18,
            pad=20
        )

        ax.set_ylabel(metrica, fontsize=13, labelpad=8)
        ax.tick_params(axis="x", labelsize=13)
        ax.tick_params(axis="y", labelsize=11)

        # OFFSET DINÂMICO DO JEITO CORRETO
        ax_ymin, ax_ymax = ax.get_ylim()
        offset = (ax_ymax - ax_ymin) * 0.01   # 3% da altura do eixo

        for bar in bars:
            val = bar.get_height()
            ax.text(
                bar.get_x() + bar.get_width()/2,
                val + offset,                # <<< SEMPRE grudado na barra
                f"{val:.2f}",
                ha="center",
                va="bottom",
                fontsize=12,
                fontweight="bold"
            )

        ax.grid(axis='y', linestyle='--', alpha=0.4)

    fig.subplots_adjust(
        top=0.92,
        bottom=0.06,
        left=0.07,
        right=0.97,
        hspace=0.40,
        wspace=0.25
    )

    with PdfPages(nome_pdf) as pdf:
        pdf.savefig(fig)

    plt.close(fig)

for posicao, pares in duelos.items():

    metricas = metricas_por_posicao[posicao]

    for jogador_fla, jogador_pal in pares:

        nome_pdf = f"duelo_{posicao}_{jogador_fla.replace(' ', '_')}_vs_{jogador_pal.replace(' ', '_')}.pdf"

        gerar_duelo_pdf(
            df_flamengo,
            df_palmeiras,
            df_gk,
            jogador_fla,
            jogador_pal,
            posicao,
            metricas,
            nome_pdf
        )

        print("PDF gerado:", nome_pdf)
```
