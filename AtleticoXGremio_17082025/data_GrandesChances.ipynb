```python
path = '/content/drive/MyDrive/data_serieA/data_gc.csv'
```

```python
import pandas as pd

data_gc = pd.read_csv(path, sep=";", decimal=",")
```

```python
data_gc.head()
```

```python
data_gc['GC'] = pd.to_numeric(data_gc['GC'], errors='coerce')
data_gc['Folha Salarial'] = pd.to_numeric(data_gc['Folha Salarial'], errors='coerce')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

cores_times = {
    "Atlético-MG": "#000000",
    "Flamengo": "#C00000",
    "Cruzeiro": "#003DA5",
    "Palmeiras": "#006437",
    "Bahia": "#004C99",
    "Botafogo": "#000000",
    "São Paulo": "#D00000",
    "Mirassol": "#F2C200",
    "Fluminense": "#8A1538",
    "Bragantino": "#E30613",  # vermelho RB
    "Ceará": "#000000",
    "Inter": "#D00000",
    "Grêmio": "#00AEEF",
    "Corinthians": "#000000",
    "Santos": "#000000",       # preto
    "Vasco": "#000000",
    "Vitória": "#D00000",
    "Juventude": "#0C8A43",
    "Fortaleza": "#0047AB",
    "Sport": "#D00000",
}

# Calculate the mean GC
mean_gc = data_gc['GC'].mean()

# ordenar times por GC
ordem = data_gc.sort_values("GC")["Time"]

# Create the bar plot
plt.figure(figsize=(15, 7))
sns.barplot(x='Time', y='GC', data=data_gc, palette=cores_times, order=ordem)

# Add the mean line
plt.axhline(mean_gc, color='red', linestyle='--', label=f'Média GC: {mean_gc:.2f}')

# Add labels and title
plt.xlabel('Time')
plt.ylabel('Grandes Chances / 90 min')
plt.title('Grandes Chances por 90 Minutos')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# eixo y de 1 até o próximo inteiro acima do máximo
plt.ylim(0.9, 2.8)

plt.show()
```
